```python
import numpy as np 
import matplotlib.pyplot as plt 

def henon_model(x, y, a=1.4, b=0.3):

    x_next = 1 - a * x ** 2 + y
    y_next = b * x
    return x_next, y_next

steps = 100000
X = np.zeros(steps + 1)
Y = np.zeros(steps + 1)


X[0], Y[0] = 0, 0

for i in range(steps):
    x_next, y_next = henon_model(X[i], Y[i])
    X[i+1] = x_next
    Y[i+1] = y_next
    

plt.plot(X, Y, '^', alpha = 0.8, markersize=0.3)
plt.title('Henon Attractor')
plt.savefig('Henon_Attractor.png')
plt.show()

plt.close()
```
